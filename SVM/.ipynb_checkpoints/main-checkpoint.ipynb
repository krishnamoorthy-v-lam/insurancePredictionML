```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
#dataset --> https://www.kaggle.com/datasets/yasserh/housing-prices-dataset
data = pd.read_csv("../DataSet/pre_processed_data.csv")
```

```python
data.head()
```

```python
x = data.loc[ : , ["age", "bmi", "children", "sex_male", "smoker_yes"] ]
```

```python
x.head()
```

```python
y = data.loc[: , ["charges"]]
```

```python
y.head()
```

# Normalization

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
```

```python
xdata = sc_x.fit_transform(x)
ydata = sc_y.fit_transform(y.values.reshape(-1, 1)).ravel()
```

```python
xdata, ydata
```

```python
x_train, x_test, y_train, y_test = train_test_split(xdata, ydata, random_state=0) # herea random state help to give same result of split when ever run the code
```

# Train a Model

```python
model = SVR(kernel="rbf", C=10)
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.support_
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

# Export model

```python
import joblib
```

```python
joblib.dump(model, "svr.joblib")
```

# Load model and predict

```python
loadedmodel = joblib.load("svr.joblib")
```

```python
inputData = sc_x.transform([[25, 34, 0, True, False]])
```

```python
res = loadedmodel.predict(inputData)
res
```
