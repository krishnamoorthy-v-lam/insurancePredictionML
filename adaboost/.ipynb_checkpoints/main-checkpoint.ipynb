```python
from sklearn.ensemble import AdaBoostRegressor
import pandas as pd
```

```python
data = pd.read_csv("../DataSet/pre_processed_data.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
x = data.loc[:, ["age", "bmi", "children", "sex_male", "smoker_yes"]]# independed variable 
```

```python
y = data[["charges"]] # depended variable
```

```python
x.head()
```

```python
x.count()
```

```python
y.head()
```

```python
y.count()
```

# Split data 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.size, x_test.size, y_train.size, y_test.size
```

# Train a model

```python
model = AdaBoostRegressor(random_state=0, learning_rate=0.1, loss="exponential")
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

# Evalute model

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

# Export Model

```python
import joblib
```

```python
joblib.dump(model, "adaboost.joblib")
```

```python
loadedmodel = joblib.load("adaboost.joblib")
```

```python
userInput = [[25, 34, 0, True, False]]
```
